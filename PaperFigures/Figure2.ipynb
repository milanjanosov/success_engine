```python
execfile("0_imports.py")
import matplotlib.pyplot as plt
import seaborn
import numpy as np
%matplotlib inline 
seaborn.set_style('white')   


  
def plot_career(inputfile, title, color, ax, title_font, labelsize, font_tick, meas = 3):
  
    data  = []
    
    for line in open(inputfile):
        fields  = line.replace(',','').replace('None', '0').strip().split('\t')                
        data.append((float(fields[1]), float(fields[meas]) ))                   
   

    years,    cnt      = zip(*sorted(data, key=lambda tup: tup[0]))
    (maxyear, maxcnt)  = sorted(data, key=lambda tup: tup[1])[-1]

    
    [ax.axvline(x = years[ij], ymin = 0, ymax = cnt[ij]/(1.15*maxcnt+0.01*maxcnt),  linewidth = 1, color = 'grey')  for ij in range(len(years))]

    ax.set_xlim([min(years)-2,max(years)+2])
    ax.set_ylim([-0.01*maxcnt,1.15*maxcnt])

    ax.plot(maxyear,  maxcnt, 'o', color = 'black', markersize = 11, alpha = 0.45, markeredgecolor='yellow', markeredgewidth = 20)#, label = title)
    ax.plot(years, cnt, 'o', color = color, markersize = 11, alpha = 0.6, markeredgecolor='grey', markeredgewidth = 2)#, linewidth='0'
    
    
    yticks = np.arange(0, 1.15*maxcnt, round(1.15*maxcnt/5, -5 if maxcnt > 500000 else -4))
    ax.set_yticks(yticks)  
    ax.set_yticklabels([str(yy/1000).replace('.0', '') + 'k' if len(str(yy).replace('.0', '')) < 7 else str(yy/1000000).replace('.0', '') + 'M' for yy in yticks])#, rotation='vertical')
    
    
    '''
    ax.set_title('The career of George Lucas', fontsize = title_font+2)
    ax.legend(loc = 'left', fontsize = font_tick)
    ticklabels = ax.get_xticklabels() + ax.get_yticklabels()
    ax.tick_params(labelsize = font_tick, pad = 9)   
    ax.set_yticklabels(['0', '0','200k', '400k', '600k', '800k', '1M'])#, rotation='vertical')
    ax.set_xlabel('Rank of movies',    fontsize = title_font)
    ax.set_ylabel('Number of ratings', fontsize = title_font)  
    #ax.xaxis.labelpad = 10
    ax.yaxis.labelpad = 10   
    '''

    
def plot_inflation(ax, label, color):
    
    (x, y, err) = zip(*[tuple(float(fff) for fff in line.strip().split('\t'))     for line in open('FinalData/2_' + label + '_inflation.dat')  ]  )
    x   = np.asarray(x)
    y   = np.asarray(y)
    err = np.asarray(err)


    ax.fill_between(x, y-err, y+err, alpha = 0.2, color = color, linewidth = 0)
    ax.plot(x, y, linewidth = 2, color = color)
    ax.set_title('Inflation curve', fontsize = 18)
    ax.set_xlabel('Time', fontsize = 18)
    ax.set_ylabel('Annual average rating', fontsize = 18)

   
    
    
def plot_impact(ax, label, color):    

    (xlog,  ylog) = zip(*[tuple(float(ijk) for ijk in line.strip().split('\t'))  for line in open('ResultData/1_impact_distributions/' + label + '_impact_distributions_lognormal_pdf_years_all.dat') ])
    (xplog, yplog) = zip(*[tuple(float(ijk) for ijk in line.strip().split('\t'))  for line in open('ResultData/1_impact_distributions/'+ label + '_impact_distributions_lognormal_hist_pdf_100_years_all.dat') ])

    

    
    ax.plot(xlog,  ylog,'-', color = color, linewidth = 2)
    ax.plot(xplog, yplog, 'o', color = color, markersize = 12, alpha = 0.5, linewidth = 0)

    
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Normalized rating count', fontsize = 18)
    ax.set_ylabel('-', fontsize = 18)
    ax.set_title('Impact distribution', fontsize = 18)



    
    
    
    
    
    
    
    
    
fig, ax = plt.subplots(3, 3, figsize=(30, 24))

plot_career('FinalData/2_example_kubrick.dat', 'Star Wars IV', 'steelblue', ax[0,0], 30, 28, 24)   
plot_career('FinalData/2_example_tiesto.dat',  'Valami', 'darkred', ax[0,1], 30, 28, 24, 2)   
plot_career('FinalData/2_example_martin.dat',  'valamike', 'darkgreen', ax[0,2], 30, 28, 24)   


plot_inflation(ax[1,0], 'director', 'steelblue') 
plot_inflation(ax[1,1], 'electro',  'darkred') 
plot_inflation(ax[1,2], 'author',   'darkgreen') 
    
    
plot_impact(ax[2,0], 'director',  'steelblue')
plot_impact(ax[2,1], 'electro',   'darkred') 
plot_impact(ax[2,2], 'authors',   'darkgreen') 

    
    
align_plot(ax)
#plt.close()
```

```python
def get_mtx(N, var, mode, ax):

    mtx    = np.zeros((N,N))  

    for line in open('ResultData/5_pQ_fit/SIM_matrix_' + var + '_' + mode + '.dat'):

        fff = line.strip().split('\t')

        ind1 = fields.index(fff[0])
        ind2 = fields.index(fff[1])
        dist = float(fff[2])

        mtx[ind1, ind2] = dist

    ax.set_title('Similarity - P(' + var + '), ' + mode + ' centered', fontsize = 28)
    ax.imshow(mtx)
  
    return mtx


N      = 14
fields = [ 'director', 'producer', 'writer', 'composer', 'art-director', 'electro', 'pop', 'rock', 'funk', 'folk', 'classical', 'jazz', 'hiphop', 'authors']
ff, ax = plt.subplots(2, 2, figsize=(24, 18))


    
mtx_pmean = get_mtx(N, 'p', 'mean', ax[1,1])
mtx_ppeak = get_mtx(N, 'p', 'peak', ax[0,1])
mtx_Qmean = get_mtx(N, 'Q', 'mean', ax[1,0])
mtx_Qpeak = get_mtx(N, 'Q', 'peak', ax[0,0])

align_plot(ax)
plt.show()
```

```python
def get_ranking(tipus = '', skew = True):

    if skew == True:
        
        labelQ, DQ, meanQ, varianceQ, skewnessQ, kurtosityQ = zip(*[[ff for ff in line.strip().split('\t') ] for line in open('ResultData/5_pQ_fit/STAT_log_Q.dat') if 'label' not in line])
        labelp, Dp, meanp, variancep, skewnessp, kurtosityp = zip(*[[ff for ff in line.strip().split('\t') ] for line in open('ResultData/5_pQ_fit/STAT_log_p.dat') if 'label' not in line])

    else:
        labelQ, DQ, meanQ, varianceQ = zip(*[[ff for ff in line.strip().split('\t') ] for line in open('ResultData/5_pQ_fit/STAT_log_Q_' + tipus + 'norm.dat') if 'label' not in line])
        labelp, Dp, meanp, variancep = zip(*[[ff for ff in line.strip().split('\t') ] for line in open('ResultData/5_pQ_fit/STAT_log_p_' + tipus + 'norm.dat') if 'label' not in line])

 
    for index, label in enumerate(labelQ):
        Qvars[label] = float(varianceQ[index])

    for index, label in enumerate(labelp):
        pvars[label] = float(variancep[index])

      
    dist_based = []
    test_based = []
    for label in labelQ:
        dist_based.append((label, (Qvars[label] - pvars[label]) / math.sqrt(2)))
        test_based.append((label, Qvars[label] / (pvars[label] + Qvars[label])))
        #print label, '\t', (Qvars[label] - pvars[label]) / math.sqrt(2) ,'\t', pvars[label] / (pvars[label] + Qvars[label])        
    
    
    dist_based.sort(key=lambda x: x[1])
    test_based.sort(key=lambda x: x[1])
   
    
    rank_dist = [e[0] for e in dist_based]  
    rank_test = [e[0] for e in test_based]  
      
    return rank_dist, rank_test


        
skew_dist,  skew_test  = get_ranking()   
norm_dist,  norm_test  = get_ranking('f', skew = False)
trans_dist, trans_test = get_ranking('t', skew = False)


from scipy.stats import spearmanr, kendalltau, pearsonr



rankings = [skew_dist, skew_test, norm_dist,  norm_test]#, trans_dist, trans_test]
rankingsn = ['skew_dist', 'skew_test', 'norm_dist',  'norm_test', 'trans_dist', 'trans_test']

correl = np.zeros((4,4))  


for r1 in rankings:
    #print r1
    for r2 in rankings:
        correl[rankings.index(r1)][rankings.index(r2)] = spearmanr(r1, r2)[0]
        print rankingsn[rankings.index(r1)], rankingsn[rankings.index(r2)], spearmanr(r1, r2)[0]
        #if spearmanr(r1, r2)[0] <0.5:
         #   print r1
         #   print r2
#        print spearmanr(r1, r2)[0]

fig, ax = plt.subplots(figsize=(6,6))
ax.imshow(correl, vmin=0, vmax=1)
ax.set_ylim([-1,4])
plt.show()

#for i in range(14):
#    print skew_dist[i], skew_test[i], norm_dist[i],  norm_test[i], trans_dist[i], trans_test[i]

    
for field in fields:
    ranks = [skew_dist.index(field), skew_test.index(field), norm_dist.index(field), norm_test.index(field), trans_dist.index(field), trans_test.index(field)]
 #   print field, np.mean(ranks), np.std(ranks)
```

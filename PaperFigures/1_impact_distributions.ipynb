```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
%matplotlib inline 

def align_plot(ax):

    ax.legend(loc = 'left', fontsize = font_tick)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ticklines  = ax.get_xticklines()  + ax.get_yticklines()
    gridlines  = ax.get_xgridlines()  + ax.get_ygridlines()
    ticklabels = ax.get_xticklabels() + ax.get_yticklabels()
    
    for line in ticklines:
        line.set_linewidth(1)

    for line in gridlines:
        line.set_linestyle('-.')

    ax.xaxis.labelpad = 15
    ax.yaxis.labelpad = 15   
    ax.tick_params(labelsize = font_tick)    



def plot_red_lines(ax, title_font, labelsize, font_tick):

    x = np.arange(0,1, 1.0/20)

    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(-0.05, 1.05)
    yyy = [1 - y0 for y0 in x]
    ax.plot(x, yyy,  '--', linewidth = 8, color =  'k', markersize = 0, alpha = 0.95, label = 'Uniform $(0,1)$') 
    ax.set_xlabel('$r$', fontsize = title_font)
    ax.set_ylabel( r'$P( \geq  r)$ for rating counts' , fontsize = title_font)




def plot_NN(ax, title_font, labelsize, font_tick):

    alpha_hist = 0.2

    (xmovie, pmovie, perrmovie) =  zip(*[ [float(a) for a in line.strip().split('\t')] for line in open('2_r_rule_data/directorNNstar_data.dat')])  
    xmovie    = np.asarray(list(xmovie)) 
    pmovie    = np.asarray(list(pmovie))
    perrmovie = np.asarray(list(perrmovie)) 
    ax.fill_between(xmovie, pmovie-perrmovie, pmovie+perrmovie, alpha = 0.2, color = 'steelblue')            
    ax.errorbar(xmovie, pmovie, yerr=perrmovie, fmt= '-', linewidth = 8, color =  'steelblue', markersize = 0, marker = 'o', alpha = 0.95, label = 'Data') 
    ax.yaxis.labelpad = 10   


    plot_red_lines(ax, title_font, labelsize, font_tick)
    align_plot(ax) 
    
    
def test_r_model(ax, title_font, labelsize, font_tick):

   
    
    (xdata, ydata) =  zip(*[ [float(a) for a in line.strip().split('\t')] for line in open('3_r_model_data/director_r_model_raw_data.dat')])      
    ax.plot(xdata, ydata, marker = 'o', markersize = 15, color = 'lightskyblue', alpha = 0.15,linewidth = 0)

    (xbdata, pbdaza) =  zip(*[ [float(a) for a in line.strip().split('\t')] for line in open('3_r_model_data/director_r_model_bin_gen.dat')])  
    ax.plot(xbdata, pbdaza, '--', linewidth = 8, color =  'k', markersize = 0, alpha = 0.95, label = 'R-model' ) 

    (xbgen, pbgen, pberrgen) =  zip(*[ [float(a) for a in line.strip().split('\t')] for line in open('3_r_model_data/director_r_model_bin_data.dat')])  
    #ax.errorbar(xbgen, pbgen, yerr=pberrgen, fmt= '-', linewidth = 4, color =  'steelblue', markersize = 0, marker = 'o', alpha = 0.95, label = 'Data') 
    ax.plot(xbgen, pbgen, '-', linewidth = 8, color =  'steelblue', alpha = 0.95, label = 'Data') 
    
    ax.set_xlabel('Career length', fontsize = title_font)
    ax.set_ylabel('Highest rated movies', fontsize = title_font)    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim([10,175])
    ax.set_ylim([1000, 100000])    

    align_plot(ax) 

       
   

def plot_career2(ax, title_font, labelsize, font_tick):
  
    years = []
    avgr  = []
    maxa  = []
    maxy  = []    
    data = []
    
    
    for line in open('nm0000184_writer_simple_career'):
        
        if 'year' not in line:
    
            line = line.replace('None', '0')        
            fields  = line.strip().split('\t')           
            try:
                movie   = fields[0]
                year    = int(fields[1])
                ratings = float(fields[2])
                count   = float(fields[3])   
                if count > 1000 and count < 900000:
                    years.append(year)
                    avgr.append(count)   
                    data.append((year, count))                   
                if count > 900000:
                    maxa.append(count)
                    maxy.append(year)  
                    data.append((year, count) )
            except:
                pass    
            
            
    print len(avgr) 

    years, avgr = zip(*sorted(data, key=lambda tup: tup[0]))
    years = range(1, len(years)+1)
    avgr  = list(avgr)

    seaborn.set_style('white')   

    [ax.axvline(x = years[ij], ymin=0, ymax=avgr[ij]/1050000,  linewidth=2, color = 'k')  for ij in range(len(years))] 

    ax.plot(years, avgr, 'o', color = 'steelblue', markersize = 23, alpha = 0.6, markeredgecolor='black', markeredgewidth = 2)#, linewidth='0'
    ax.plot(4,  maxa, 'o', color = 'tomato', markersize = 26, alpha = 0.85, markeredgecolor='black', markeredgewidth = 2, label = 'Star Wars IV')#, linewidth='0'
    ax.set_ylim([-20000,1050000])

    ax.set_title('The career of George Lucas', fontsize = title_font+2)
    ax.legend(loc = 'left', fontsize = font_tick)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ticklines  = ax.get_xticklines()  + ax.get_yticklines()
    gridlines  = ax.get_xgridlines()  + ax.get_ygridlines()
    ticklabels = ax.get_xticklabels() + ax.get_yticklabels()
    ax.tick_params(labelsize = font_tick, pad = 9)   
  
    ax.set_yticklabels(['0', '0','200k', '400k', '600k', '800k', '1M'])#, rotation='vertical')
    ax.set_xlabel('Rank of movies',    fontsize = title_font)
    ax.set_ylabel('Number of ratings', fontsize = title_font)  
    #ax.xaxis.labelpad = 10
    ax.yaxis.labelpad = 10   
    [line.set_linestyle('-.') for line in gridlines]
   









title_font = 36 
labelsize  = 30
font_tick  = 25



fig =plt.figure(figsize=(24,16))
fig.subplots_adjust(bottom=0.025, left=0.025, top = 0.975, right=0.975)
X = [ (2,3,(1,3)), (2,3,4), (2,2,4) ]
seaborn.set_style('white')  

import string
for index, (nrows, ncols, plot_number) in enumerate(X):

    sub = fig.add_subplot(nrows, ncols, plot_number)

    if index == 0: 
        plot_career2(sub, title_font, labelsize, font_tick)
        sub.text(-0.1, 1.1, string.ascii_lowercase[index] + ')', transform=sub.transAxes, size=34, weight='bold')       
    if index == 1: 
        plot_NN(sub, title_font, labelsize, font_tick)
        sub.text(-0.35, 1.05, string.ascii_lowercase[index] + ')', transform=sub.transAxes, size=34, weight='bold')    
    elif index == 2: 
        test_r_model(sub, title_font, labelsize, font_tick)
        sub.text(0.95, 1.00, string.ascii_lowercase[index] + ')', transform=sub.transAxes, size=34, weight='bold')    
    
       
 
    
plt.gcf().subplots_adjust(left=0.1, bottom = 0.08, top = 0.9)
plt.savefig('0_netsci_abstract.png')
plt.show()


```

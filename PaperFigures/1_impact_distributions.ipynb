```python
execfile("0_imports.py")
import matplotlib.pyplot as plt
import seaborn
import numpy as np
%matplotlib inline 
seaborn.set_style('white')       




def get_comparing_plot(fields, ax, norm):

    title_font = 42 
    labelsize = 37 
    
    for field in fields:
        (xlog, ylog) = zip(*[tuple(float(ijk) for ijk in line.strip().split('\t'))  for line in open('ResultData/1_impact_distributions/' + field + '_impact_distributions_lognormal_pdf_' + norm + '.dat') ])   
        ax.plot(xlog,ylog,'-', linewidth = 2, label = field)

    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Normalized rating count', fontsize = 18)
    ax.set_ylabel('CDF', fontsize = 18)

    ax.set_title(norm, fontsize = 18)




    
    
fig1, ax = plt.subplots(2, 2, figsize=(22, 16))
fields   = ['pop','director', 'producer', 'writer','funk', 'folk', 'classical', 'pop', 'jazz', 'rock', 'authors', 'electro', 'hiphop' ]

get_comparing_plot(fields, ax[0,0], 'no')  
get_comparing_plot(fields, ax[0,1], 'years_all')  
get_comparing_plot(fields, ax[1,0], 'field_avg')  
get_comparing_plot(fields, ax[1,1], 'fields_all')  

    
    
    

align_plot(ax)
[ax[i,j].legend(loc = 'left', fontsize = 12) for i in range(2) for j in range(2)]

plt.show()
```
